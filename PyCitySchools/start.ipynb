```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load, index_col=0)
student_data = pd.read_csv(student_data_to_load, index_col=0)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
unique_schools=len(pd.unique(school_data_complete['school_name']))
unique_schools
```

```python
total_students=len(student_data)
total_students
```

```python
total_budget=sum(school_data_complete['budget'])
total_budget
```

```python
reading_avg=sum(school_data_complete['reading_score'])/len(school_data_complete['reading_score'])
reading_avg
```

```python
math_avg=sum(school_data_complete['math_score'])/len(school_data_complete['math_score'])
math_avg
```

```python
passed_reading=school_data_complete.loc[school_data_complete['reading_score']>=70]
percent_passed_reading=len(passed_reading)/total_students*100
percent_passed_reading
```

```python
school_data['budget_per_student'] = school_data.apply(lambda x: x['budget']/x['size'], axis=1)
```

```python
passed_math=school_data_complete.loc[school_data_complete['math_score']>=70]
percent_passed_math=len(passed_math)/total_students*100
percent_passed_math
```

```python
passed_both=school_data_complete.loc[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)]
percent_passed_both=len(passed_both)/total_students*100
percent_passed_both
```

```python
district_dictionary={'Total Number of Schools':[unique_schools],'Total Number of Students':[total_students],'Total Budget':[total_budget], 'Average Math Score':[math_avg], 'Average Reading Score':[reading_avg], 'Percent Passing Math':[percent_passed_math], 'Percent Passing Reading':[percent_passed_reading], 'Percent Passing Both':[percent_passed_both]}
District_Summary=pd.DataFrame(district_dictionary)
District_Summary
```

```python
school_stats = student_data.groupby(["school_name"]).agg(avg_math=pd.NamedAgg("math_score", 'mean'),
                                        avg_reading=pd.NamedAgg("reading_score", 'mean'),
                                        pct_passing_math=pd.NamedAgg("math_score", lambda x: np.sum(x>=70)/len(x)),
                                         pct_passing_reading=pd.NamedAgg("reading_score", lambda x: np.sum(x>=70)/len(x)))
school_stats.head()
```

```python
school_stats['pass_both'] = student_data.groupby("school_name").apply(lambda x: np.sum((x['math_score']>=70)&(x['reading_score']>=70))/len(x))
```

```python
per_school_summary=school_data.merge(school_stats, on='school_name', how='outer')
per_school_summary
```

```python
school_data_complete['passed']=school_data_complete.apply(lambda x: x.reading_score >= 70 and x.math_score >= 70, axis=1)
```

```python
school_aggregate=school_data_complete.groupby(['school_name']).agg(pct_passed=pd.NamedAgg("passed", aggfunc='sum'), 
                                                 total_students=pd.NamedAgg("passed", aggfunc='count'))
```

```python
#school_data_complete.groupby(['school_name'])['passed'].apply(lambda x: x.sum()/len(x)).sort_values(ascending=False).head()
```

```python
school_aggregate['percent_pass']=school_aggregate['pct_passed']/school_aggregate['total_students']*100
school_aggregate
```

```python
top_schools=school_aggregate.sort_values('percent_pass', ascending=False).head()
top_schools
```

```python
bottom_schools=school_aggregate.sort_values('percent_pass', ascending=True).head()
bottom_schools
```

```python
schools_compiled=school_data_complete.groupby(['school_name', 'grade'])
math_scores_by_grade=schools_compiled[['math_score']].aggregate('mean')
math_scores_by_grade
```

```python
reading_scores_by_grade=schools_compiled[['reading_score']].aggregate('mean')
reading_scores_by_grade
```

```python
school_data['spending_per_student']=school_data['budget']/school_data['size']
school_data
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
per_school_summary['budget_bin'] = pd.cut(per_school_summary['budget_per_student'], spending_bins, labels=labels)
per_school_summary.head()
```

```python
spending_summary=per_school_summary.groupby(['budget_bin'])[['avg_math',
                                            'avg_reading',
                                            'pct_passing_math',
                                            'pct_passing_reading',
                                            'pass_both']].agg('mean')
spending_summary
```

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
per_school_summary['size_bins']=pd.cut(per_school_summary['size'], size_bins, labels=labels)
per_school_summary
```

```python
size_summary=per_school_summary.groupby(['size_bins'])[['avg_math',
                                            'avg_reading',
                                            'pct_passing_math',
                                            'pct_passing_reading',
                                            'pass_both']].agg('mean')
size_summary
```

```python
type_summary=per_school_summary.groupby(['type'])[['avg_math',
                                            'avg_reading',
                                            'pct_passing_math',
                                            'pct_passing_reading',
                                            'pass_both']].agg('mean')
type_summary
```
